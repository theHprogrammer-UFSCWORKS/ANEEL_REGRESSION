##### Por que escolhemos esse dataset?

A escolha desse dataset para análise focada nos agentes da CELESC baseia-se na relevância estratégica desta empresa no contexto do fornecimento de energia elétrica. A análise promete desdobrar as características intrínsecas das interrupções de energia, possibilitando uma compreensão aprofundada das incidências, suas causas e impactos. Com isso, ele pode identificar padrões e tendências que são fundamentais para aprimorar a gestão e a resposta a interrupções, minimizando o tempo de inatividade e melhorando a satisfação do consumidor.

A importância deste estudo reside na sua capacidade de influenciar positivamente a eficiência operacional. Ao desvendar os tipos de interrupções mais comuns e os motivos de expurgos, ele proporciona uma base de dados sólida para que a CELESC possa implementar medidas preventivas e corretivas mais eficazes.

Para os visualizadores da análise, os insights gerados têm o potencial de elevar a transparência operacional e a confiança. Os consumidores e reguladores poderão observar a proatividade da empresa em lidar com questões críticas e seu compromisso com a melhoria contínua.

Além disso, com a inclusão de uma regressão linear, abre-se a possibilidade de antever situações futuras de interrupções, baseando-se nas variáveis históricas do dataset. Esta modelagem preditiva poderia ser uma ferramenta valiosa para planejamento e alocação de recursos, levando a uma resposta mais ágil em eventos de interrupção e, em última instância, a um fornecimento de energia mais estável e confiável.

7. Identificação de Pontos Críticos:

```python
# Valores nulos em DscInterrupcaocausa e DscInterrupcaoDescricao, remover a linha
df_analise.dropna(subset=['DscInterrupcaoCausa', 'DscInterrupcaoDescricao'], inplace=True)

# Remover a coluna DscFatoGeradorInterrupcao
df_analise.drop('DscFatoGeradorInterrupcao', axis=1, inplace=True)

df_analise.info()
```

```python
# Mudar todos e tudo o que contem "Não" para "Nao"
df_analise.replace(to_replace="Não Programada", value="Nao programada", inplace=True) 
df_analise.replace(to_replace="Não Classificada", value="Nao classificada", inplace=True)
df_analise.replace(to_replace="Não Informada", value="Nao informada", inplace=True)
df_analise.replace(to_replace="Não Identificada", value="Nao identificada", inplace=True)
```

```python
# Remover DscTipoInterrupcao
df_analise.drop('DscTipoInterrupcao', axis=1, inplace=True)
```

```python
# Criar uma nova coluna chamada DatSemana contendo se a DatInicioInterrupcao é 'Dia de semana' ou 'Fim de semana'
df_analise['DatSemana'] = np.where(pd.to_datetime(df_analise['DatInicioInterrupcao']).dt.dayofweek < 5, 'Dia de semana', 'Fim de semana')

df_analise.head()
```

```python
# Criar uma nova coluna chamada DatFeriado contedo False ou True se a DatInicioInterrupcao é feriado ou não
holiday_list = []
for holiday in holidays.Brazil(years=[2023]).items():
    holiday_list.append(holiday)
    
holidays_df = pd.DataFrame(holiday_list, columns=["date", "holiday"])
pd.to_datetime(holidays_df['date'])

df_analise['DatFeriado'] = np.where(pd.to_datetime(df_analise['DatInicioInterrupcao']).dt.date.isin(holidays_df['date']), True, False)

df_analise.head()
```

```python
# Cálculo do tempo total de interrupção em minutos
df_analise['TempoTotalInterrupcao'] = (pd.to_datetime(df_analise['DatFimInterrupcao']) - pd.to_datetime(df_analise['DatInicioInterrupcao'])).dt.total_seconds() / 60

df_analise.head()
```

```python
# 3. Gráfico de Pareto considerando os valores em DscInterrupcaoCausa
# Descrição do por que fazer essa análise:


# Criar um DataFrame com o número de ocorrências por causa de interrupção
df_interrupcao_causa = df_analise.groupby('DscInterrupcaoCausa')['DscInterrupcaoCausa'].count().to_frame('Ocorrencias').reset_index()

# Ordenar o DataFrame por Ocorrencias em ordem decrescente
df_interrupcao_causa = df_interrupcao_causa.sort_values(by='Ocorrencias', ascending=False)

# Calcular a porcentagem de ocorrências
df_interrupcao_causa['Percentual'] = df_interrupcao_causa['Ocorrencias'] / df_interrupcao_causa['Ocorrencias'].sum() * 100

# Calcular a porcentagem acumulada
df_interrupcao_causa['PercentualAcumulado'] = df_interrupcao_causa['Percentual'].cumsum()

# Plotar o gráfico
sns.set_theme(style="whitegrid")
plt.figure(figsize=(20, 10))
ax = sns.barplot(x="DscInterrupcaoCausa", y="Ocorrencias", data=df_interrupcao_causa)
ax.set_title("Número de ocorrências por causa de interrupção", fontsize=16)
ax.set_xlabel("Causa de interrupção", fontsize=14)
ax.set_ylabel("Ocorrências", fontsize=14)

# Eixo secundário
ax2 = ax.twinx()

# Cria a Curva de Pareto no eixo secundário
ax2.plot(df_interrupcao_causa["DscInterrupcaoCausa"], df_interrupcao_causa["PercentualAcumulado"], color="red", marker="D", ms=7, label='Pareto')

# Configurações dos eixos
ax.tick_params(axis="y", colors="C0")
ax2.tick_params(axis="y", colors="C1")
ax2.set_ylim([0,110])

# Ajusta a inclinação dos labels do eixo x
for tick in ax.get_xticklabels():
    tick.set_rotation(45)
    
# Finalmente mostra o grafico definido acima
plt.legend() # para mostrar a legenda
plt.show()
```

##### Aplicando Clusterização

A clusterização em um dataset como o apresentado oferece várias vantagens analíticas e operacionais, principalmente ao lidar com interrupções de energia. Ao concatenar colunas como IdeMotivoInterrupcao, NumNivelTensao, NumUnidadeConsumidora, DscInterrupcaoTipo, DscInterrupcaoCausa, DscInterrupcaoDescricao, DatSemana, DatFeriado e Intervalo para criar embeddings, pode-se transformar esses dados categóricos e numéricos em representações densas de menor dimensão que capturam a estrutura subjacente dos dados e as relações não lineares entre as características.

Essas representações, ou embeddings, permitem que algoritmos de clusterização identifiquem padrões e agrupamentos que podem não ser imediatamente aparentes em uma análise tradicional. Por exemplo, a clusterização pode revelar grupos de interrupções de energia que ocorrem devido a causas semelhantes ou sob condições similares. Essa abordagem pode fornecer insights sobre as seguintes questões:

1. Padrões de Falhas: Identificar clusters de falhas similares pode ajudar a entender as condições comuns que levam a interrupções, possibilitando ações preventivas mais direcionadas.

2. Manutenção e Substituição de Equipamentos: Agrupar unidades consumidoras com base em interrupções frequentes pode indicar a necessidade de manutenção ou substituição de equipamentos em certas áreas.

3. Planejamento de Recursos: Reconhecer padrões temporais de interrupções, como aquelas que ocorrem durante feriados ou finais de semana, pode ajudar a alocar recursos de maneira mais eficaz.

4. Resposta a Emergências: Compreender a tipologia das interrupções pode melhorar a resposta a emergências e a comunicação com os clientes.

5. Melhoria Contínua: A análise de clusters pode levar a melhorias contínuas em processos e infraestrutura ao revelar as causas raízes das interrupções.

A clusterização também pode ser usada para segmentar clientes ou áreas geográficas com base em características de interrupção, permitindo uma abordagem mais personalizada para melhorias de serviço e estratégias de comunicação. Ao utilizar técnicas de aprendizado de máquina para reduzir a dimensionalidade e extrair embeddings significativos, as empresas de serviços públicos podem obter uma visão mais profunda de seus dados operacionais e impulsionar decisões baseadas em dados para melhorar a confiabilidade e a satisfação do cliente.

```python
# Método para criar o documento que será inserido no banco
def create_document(id: int, interrupcao: str):
    document = {
        "_id": id,
        "interrupcao": interrupcao,
        "embedding": embedding_model.encode(interrupcao)
    }
    return document
```

```python
# Cria uma lista de documentos a partir do DataFrame
documents = []
try:
    for i, row in df_analise_cluster.iterrows():
        # Constrói o documento usando a função build_doc
        document = create_document(row['id'], row['DscInterrupcao'])
        documents.append(document)
except:
    print('Erro ao criar a lista de documentos')

# Gera os embeddings para o conjunto de documentos
embeddings = [doc['embedding'] for doc in documents]
```

```python
# Acrescenta os embeddings ao DataFrame
df_analise_cluster['embedding'] = embeddings
```

```python
# Converte os embeddings para uma matriz numpy
embeddings_matrix = np.stack(df_analise_cluster['embedding'].values)

# Reduz a dimensionalidade dos embeddings para 3 com o PCA
pca = PCA(n_components=3)
reduced_embeddings = pca.fit_transform(embeddings_matrix)

# Cria um DataFrame com os embeddings reduzidos
df_pca = pd.DataFrame(reduced_embeddings, columns=['x', 'y', 'z'])

# Acrescenta as coordenadas do PCA ao DataFrame
df_analise_cluster['x'] = reduced_embeddings[:, 0]
df_analise_cluster['y'] = reduced_embeddings[:, 1]
df_analise_cluster['z'] = reduced_embeddings[:, 2]

# Mostra as primeiras linhas do DataFrame
df_analise_cluster.head()
```

A plotagem tridimensional após a redução de dimensionalidade via PCA (Análise de Componentes Principais) revela uma distribuição espacial interessante dos dados, onde pode-se observar agrupamentos distintos que são provavelmente representativos de diferentes tipos de interrupções de energia ou condições associadas a elas.

```python
# Usando o método Calinski-Harabasz Index
try:
    print('Calinski-Harabasz Index')
    ch_scores = []
    for i in range(2, 11):
        kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
        kmeans.fit(df_pca)
        ch_score = metrics.calinski_harabasz_score(df_pca, kmeans.labels_)
        ch_scores.append(ch_score)
        print(f"Clusters: {i} - Calinski-Harabasz Score: {ch_score}")
except:
    print('Erro ao executar o Calinski-Harabasz Index')
```

A metodologia do cotovelo indicou um k (número de clusters) igual a 4 como o ponto ótimo para a clusterização, sugerindo que quatro grupos distintos poderiam ser identificados nesse espaço de características. No entanto, o Índice Calinski-Harabasz, que é uma medida de validação de cluster que avalia a dispersão entre clusters e a dispersão dentro dos clusters, sugere que o número ótimo de clusters seria 5, já que esse número de clusters produziu a pontuação mais alta, indicando agrupamentos mais definidos e melhor separação entre eles.

Essa discrepância entre os métodos de determinação de k indica que, embora o método do cotovelo sugira uma solução com 4 clusters, a análise de Calinski-Harabasz, que considera a coesão interna dos clusters e a separação entre eles, recomenda um cluster adicional para capturar melhor a estrutura dos dados. Isso pode ser devido à natureza dos dados ou às particularidades de como os clusters se formam no espaço de características reduzido.

Com um k escolhido de 5, a análise de cluster pode proceder com a expectativa de que cinco grupos distintos de interrupções serão identificados, cada um representando diferentes combinações de motivos de interrupção, níveis de tensão, tipos, causas, descrições de interrupções e contextos temporais (como dia da semana e feriados). O cluster sugere que, embora as manutenções programadas sejam geralmente bem gerenciadas, ainda existem desafios significativos com interrupções não programadas. A presença de interrupções devido a defeitos internos e objetos na rede indica a necessidade de uma vigilância e resposta rápidas para evitar e mitigar falhas. Além disso, a frequência de eventos relacionados a terceiros implica que poderia haver uma oportunidade de melhorar a segurança e a proteção da infraestrutura, bem como a comunicação com o público e as partes interessadas para reduzir esses incidentes. A análise detalhada das causas e durações das interrupções neste cluster pode fornecer informações valiosas para o planejamento de infraestrutura e resposta a emergências, visando aprimorar a resiliência do sistema e a satisfação do cliente. A implementação de tecnologias avançadas de monitoramento e diagnóstico, juntamente com a educação pública sobre os riscos de interferência na rede elétrica, são medidas potenciais que podem resultar de tais análises.

Este padrão destaca a influência significativa do ambiente externo sobre a confiabilidade do serviço de energia. Os dados refletem um desafio comum para as empresas de energia: gerenciar e mitigar os riscos ambientais que podem causar interrupções de serviço. As interrupções causadas por árvores ou vegetação geralmente requerem poda ou remoção, enquanto as interrupções por animais podem necessitar de barreiras ou medidas de proteção para evitar danos. A frequência de eventos relacionados à vegetação e animais sugere que uma gestão proativa do ambiente ao redor das linhas de transmissão e equipamentos pode ser necessária para reduzir o número de falhas. Isso pode incluir programas de manutenção de vegetação mais rigorosos e a instalação de dispositivos para desencorajar animais de se aproximarem das instalações elétricas. Por outro lado, as descargas atmosféricas, como raios, representam um tipo de interrupção que é mais difícil de prever e prevenir. A implementação de sistemas de proteção contra raios e o aprimoramento da infraestrutura para resistir a tais eventos podem ser estratégias úteis.
