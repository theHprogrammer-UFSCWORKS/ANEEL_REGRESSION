##### Por que escolhemos esse dataset?

A escolha desse dataset para análise focada nos agentes da CELESC baseia-se na relevância estratégica desta empresa no contexto do fornecimento de energia elétrica. A análise promete desdobrar as características intrínsecas das interrupções de energia, possibilitando uma compreensão aprofundada das incidências, suas causas e impactos. Com isso, ele pode identificar padrões e tendências que são fundamentais para aprimorar a gestão e a resposta a interrupções, minimizando o tempo de inatividade e melhorando a satisfação do consumidor.

A importância deste estudo reside na sua capacidade de influenciar positivamente a eficiência operacional. Ao desvendar os tipos de interrupções mais comuns e os motivos de expurgos, ele proporciona uma base de dados sólida para que a CELESC possa implementar medidas preventivas e corretivas mais eficazes.

Para os visualizadores da análise, os insights gerados têm o potencial de elevar a transparência operacional e a confiança. Os consumidores e reguladores poderão observar a proatividade da empresa em lidar com questões críticas e seu compromisso com a melhoria contínua.

Além disso, com a inclusão de uma regressão linear, abre-se a possibilidade de antever situações futuras de interrupções, baseando-se nas variáveis históricas do dataset. Esta modelagem preditiva poderia ser uma ferramenta valiosa para planejamento e alocação de recursos, levando a uma resposta mais ágil em eventos de interrupção e, em última instância, a um fornecimento de energia mais estável e confiável.

1. Tipos de Interrupção:

4. Impacto no Consumidor:

    - Investigar a relação entre os motivos de expurgo e os fatos geradores de interrupções.

6. Relação entre Nível de Tensão e Interrupções:
    - Estudar como diferentes níveis de tensão estão relacionados com a frequência e duração das interrupções.

7. Identificação de Pontos Críticos:
    - Localizar áreas geográficas ou alimentadores específicos que apresentam maior número de falhas.

8. Previsão de Interrupções:

```python
# Remover linhas duplicadas
df_analise.drop_duplicates(inplace=True)

# Verifica se tem valores duplicados
print(f"Tem valores duplicados? {df_analise.duplicated().any()}")
```

```python
# Remover colunas que não serão utilizadas
drop = ['_id', 'DatGeracaoConjuntoDados', 'NumOrdemInterrupcao', 'NumAno', 'NomAgenteRegulado', 'NumCPFCNPJ', 'rank']
df_analise.drop(drop, axis=1, inplace=True)

# Indice de colunas
df_analise.columns
```

```python
# Remover Agentes que não são da CELESC
df_analise = df_analise[df_analise['SigAgente'].str.contains("CELESC")]

# Mostrar a quantidade de registros
print(f"Quantidade de registros: {df_analise.shape[0]}")
print(f"Agentes: {df_analise['SigAgente'].unique()}")
```

```python
# Verificar se tem valores nulos
print(f"Tem valores nulos? {df_analise.isnull().values.any()}")
```

```python
# Mostrar lista de valores em DscFatoGeradorInterrupcao
print("Valores em DscFatoGeradorInterrupcao:")
display(df_analise['DscFatoGeradorInterrupcao'].unique())
```

```python
# Separar os valores da coluna DscFatoGeradorInterrupcao
df_analise[['DscInterrupcaoOrigem', 'DscInterrupcaoTipo', 'DscInterrupcaoCausa', 'DscInterrupcaoDescricao']] = df_analise['DscFatoGeradorInterrupcao'].str.split("/", expand=True)

df_analise.info()
```

```python
# Valores nulos em DscInterrupcao3 e DscInterrupcao4, filtrar para Não Classificado (fillna)
df_analise['DscInterrupcaoCausa'].fillna("Nao Classificada", inplace=True)
df_analise['DscInterrupcaoDescricao'].fillna("Nao Classificada", inplace=True)

# Remover a coluna DscFatoGeradorInterrupcao
df_analise.drop('DscFatoGeradorInterrupcao', axis=1, inplace=True)

df_analise.info()
```

```python
# Mudar todos e tudo o que contem "Nao" para "Não"
df_analise.replace(to_replace="Não Programada", value="Nao programada", inplace=True) 
df_analise.replace(to_replace="Não Classificada", value="Nao classificada", inplace=True)
df_analise.replace(to_replace="Não Informada", value="Nao informada", inplace=True)
df_analise.replace(to_replace="Não Identificada", value="Nao identificada", inplace=True)
```

```python
# Verificar se a coluna DscInterrupcao2 é igual a DscTipoInterrupcao
print(f"Coluna DscInterrupcaoTipo é igual a DscTipoInterrupcao? {df_analise['DscInterrupcaoTipo'].equals(df_analise['DscTipoInterrupcao'])}")

# mostrar a quantidade de valores em DscInterrupcao2 que são iguais a DscTipoInterrupcao
print(f"Quantidade de valores em DscInterrupcaoTipo que são iguais a DscTipoInterrupcao: {df_analise[df_analise['DscInterrupcaoTipo'] == df_analise['DscTipoInterrupcao']].shape[0]}")
```

```python
# Lista com os valores de DscInterrupcaoTipo
print("Valores em DscInterrupcaoTipo:")
display(df_analise['DscInterrupcaoTipo'].unique())

# Lista com os valores de DscTipoInterrupcao
print("Valores em DscTipoInterrupcao:")
display(df_analise['DscTipoInterrupcao'].unique())
```

```python
# Criar uma nova coluna chamada DatSemana contendo se a DatInicioInterrupcao é 'Dia de semana' ou 'Fim de semana'
df_analise['DatSemana'] = np.where(pd.to_datetime(df_analise['DatInicioInterrupcao']).dt.dayofweek < 5, 'Dia de semana', 'Fim de semana')

df_analise.head()
```

```python
# Criar uma nova coluna chamada DatFeriado contedo False ou True se a DatInicioInterrupcao é feriado ou não
holiday_list = []
for holiday in holidays.Brazil(years=[2023]).items():
    holiday_list.append(holiday)
    
holidays_df = pd.DataFrame(holiday_list, columns=["date", "holiday"])
pd.to_datetime(holidays_df['date'])

df_analise['DatFeriado'] = np.where(pd.to_datetime(df_analise['DatInicioInterrupcao']).dt.date.isin(holidays_df['date']), True, False)

df_analise.head()
```

```python
# Cálculo do tempo total de interrupção em minutos
df_analise['TempoTotalInterrupcao'] = (pd.to_datetime(df_analise['DatFimInterrupcao']) - pd.to_datetime(df_analise['DatInicioInterrupcao'])).dt.total_seconds() / 60

df_analise.head()
```

```python
# 3. Gráfico de Pareto considerando os valores em DscInterrupcaoCausa
# Descrição do por que fazer essa análise:


# Criar um DataFrame com o número de ocorrências por causa de interrupção
df_interrupcao_causa = df_analise.groupby('DscInterrupcaoCausa')['DscInterrupcaoCausa'].count().to_frame('Ocorrencias').reset_index()

# Ordenar o DataFrame por Ocorrencias em ordem decrescente
df_interrupcao_causa = df_interrupcao_causa.sort_values(by='Ocorrencias', ascending=False)

# Calcular a porcentagem de ocorrências
df_interrupcao_causa['Percentual'] = df_interrupcao_causa['Ocorrencias'] / df_interrupcao_causa['Ocorrencias'].sum() * 100

# Calcular a porcentagem acumulada
df_interrupcao_causa['PercentualAcumulado'] = df_interrupcao_causa['Percentual'].cumsum()

# Plotar o gráfico
sns.set_theme(style="whitegrid")
plt.figure(figsize=(20, 10))
ax = sns.barplot(x="DscInterrupcaoCausa", y="Ocorrencias", data=df_interrupcao_causa)
ax.set_title("Número de ocorrências por causa de interrupção", fontsize=16)
ax.set_xlabel("Causa de interrupção", fontsize=14)
ax.set_ylabel("Ocorrências", fontsize=14)

# Eixo secundário
ax2 = ax.twinx()

# Cria a Curva de Pareto no eixo secundário
ax2.plot(df_interrupcao_causa["DscInterrupcaoCausa"], df_interrupcao_causa["PercentualAcumulado"], color="red", marker="D", ms=7, label='Pareto')

# Configurações dos eixos
ax.tick_params(axis="y", colors="C0")
ax2.tick_params(axis="y", colors="C1")
ax2.set_ylim([0,110])

# Ajusta a inclinação dos labels do eixo x
for tick in ax.get_xticklabels():
    tick.set_rotation(45)
    
# Finalmente mostra o grafico definido acima
plt.legend() # para mostrar a legenda
plt.show()
```

```python
# 4. Tabela de frequência Causa x Descrição
# Descrição do por que fazer essa análise:

# Criar um DataFrame com o número de ocorrências por causa e descrição de interrupção
df_causa_descricao = df_analise.groupby(['DscInterrupcaoCausa', 'DscInterrupcaoDescricao'])['DscInterrupcaoDescricao'].count().to_frame('Ocorrencias').reset_index()

# Ordenar o DataFrame por Ocorrencias em ordem decrescente
df_causa_descricao = df_causa_descricao.sort_values(by='Ocorrencias', ascending=False)

df_causa_descricao
```

```python
# Usar o método pivot para criar uma tabela de frequência com os dados do df_causa_descricao

# Criar o pivot
df_causa_descricao_pivot = df_causa_descricao.pivot(index='DscInterrupcaoDescricao', columns='DscInterrupcaoCausa', values='Ocorrencias')

# Ordenar as colunas em ordem decrescente
df_causa_descricao_pivot = df_causa_descricao_pivot.reindex(df_causa_descricao_pivot.sum().sort_values(ascending=False).index, axis=1)

# Ordenar as linhas em ordem decrescente
df_causa_descricao_pivot = df_causa_descricao_pivot.reindex(df_causa_descricao_pivot.sum(axis=1).sort_values(ascending=False).index)

# Filtrar os valores nulos e substituir por 0
df_causa_descricao_pivot = df_causa_descricao_pivot.fillna(0)

# Mostrar a tabela de frequência com mapas de calor
sns.set_theme(style="whitegrid")
plt.figure(figsize=(20, 10))
ax = sns.heatmap(df_causa_descricao_pivot, annot=True, fmt=".0f", cmap="Blues")

plt.show()
```

```python
# 3. Duração das Interrupções:
#     - Calcular a duração média das interrupções e identificar as de maior impacto.

# Criar um DataFrame com a duração média das interrupções por tipo e causa de interrupção
df_duracao_media = df_analise.groupby(['DscTipoInterrupcao', 'DscInterrupcaoCausa'])['TempoTotalInterrupcao'].mean().to_frame('DuracaoMedia').reset_index()

# Ordenar o DataFrame por tipo e depois por DuracaoMedia em ordem decrescente
df_duracao_media = df_duracao_media.sort_values(by=['DscTipoInterrupcao', 'DuracaoMedia'], ascending=[True, False])

# Mostrar o DataFrame
df_duracao_media
```

```python
# Plotar o gráfico
sns.set_theme(style="whitegrid")
plt.figure(figsize=(20, 10))
ax = sns.barplot(x="DuracaoMedia", y="DscInterrupcaoCausa", hue="DscTipoInterrupcao", data=df_duracao_media)
ax.set_title("Duração média das interrupções por tipo e causa de interrupção", fontsize=16)
ax.set_xlabel("Duração média", fontsize=14)
ax.set_ylabel("Causa de interrupção", fontsize=14)
plt.show()
```

```python
# 4. Análise Temporal:
#     - Observar tendências ao longo do tempo para identificar padrões sazonais ou anuais de interrupções.

# Criar um DataFrame com o número de ocorrências por ano e mês (Pegar mês e ano de DatInicioInterrupcao)
df_ano_mes = df_analise.groupby([pd.to_datetime(df_analise['DatInicioInterrupcao']).dt.strftime('%Y-%m')])['DatInicioInterrupcao'].count().to_frame('Ocorrencias').reset_index()

# Ordenar o DataFrame por ano e mês
df_ano_mes = df_ano_mes.sort_values(by='DatInicioInterrupcao', ascending=True)

# Mostrar o DataFrame
df_ano_mes
```

```python
# Mostrar o gráfico
sns.set_theme(style="whitegrid")
plt.figure(figsize=(20, 10))
ax = sns.lineplot(x="DatInicioInterrupcao", y="Ocorrencias", data=df_ano_mes)
ax.set_title("Número de ocorrências por ano e mês", fontsize=16)
ax.set_xlabel("Ano e mês", fontsize=14)
ax.set_ylabel("Ocorrências", fontsize=14)
plt.show()
```

```python
# 5. Impacto no Consumidor:
#     - Avaliar o número de unidades consumidoras afetadas por interrupções e o tempo médio de restabelecimento do serviço.

# Criar um DataFrame com o número de ocorrências por tipo e causa de interrupção
df_impacto_consumidor = df_analise.groupby(['DscTipoInterrupcao', 'DscInterrupcaoCausa'])['NumUnidadeConsumidora'].sum().to_frame('UnidadesConsumidoras').reset_index()

# Ordenar o DataFrame por tipo e depois por UnidadesConsumidoras em ordem decrescente
df_impacto_consumidor = df_impacto_consumidor.sort_values(by=['DscTipoInterrupcao', 'UnidadesConsumidoras'], ascending=[True, False])

# Mostrar o DataFrame
df_impacto_consumidor
```

```python
# Mostrar o gráfico
sns.set_theme(style="whitegrid")
plt.figure(figsize=(20, 10))
ax = sns.barplot(x="UnidadesConsumidoras", y="DscInterrupcaoCausa", hue="DscTipoInterrupcao", data=df_impacto_consumidor)
ax.set_title("Número de unidades consumidoras afetadas por interrupções por tipo e causa de interrupção", fontsize=16)
ax.set_xlabel("Unidades consumidoras", fontsize=14)
ax.set_ylabel("Causa de interrupção", fontsize=14)
plt.show()
```
